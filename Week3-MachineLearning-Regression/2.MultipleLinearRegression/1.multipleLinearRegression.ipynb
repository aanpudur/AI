```python
# Install pandas if not already installed
%pip install pandas

import pandas as pd

df = pd.read_csv("50_Startups.csv")
df
```

```python
df = pd.get_dummies(df,dtype=int, drop_first=True)
```

```python
df
```

```python
X = df[["R&D Spend", "Administration",	"Marketing Spend", "State_Florida", "State_New York"]]
```

```python
X
```

```python
#Dependant variable assign to y

y = df[["Profit"]]
```

```python
y
```

```python
#splitting the data using the sklearn.model selection library using train_test_split and assign the value to X_train, y_train, X_test, y_test
from sklearn.model_selection import train_test_split

X_train,  X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

#create and train model
model = LinearRegression()
model.fit(X_train, y_train)

#get bias - intercept
bias = model.intercept_
bias
```

```python
#getting the slope 
weight = model.coef_
```

```python
weight
```

```python
#making prediction

y_pred = model.predict(X_test)

```

```python
from sklearn.metrics import r2_score
#evaluate the prediction
r_score = r2_score(y_test, y_pred)
r_score
```

```python
# saving file

import pickle
model_fname = "final_Mult_linear.sav"
with open(model_fname,'wb') as file:
    pickle.dump(model, file)

```

```python
with open(model_fname,'rb') as file:
    loaded_model = pickle.load(file)
```

```python


# Get feature names from the original training data
feature_names = X_train.columns.tolist()

# Create a DataFrame with the same feature names
sample_input = pd.DataFrame([[16449.20, 16897.80, 41784.10, 0, 1]], columns=feature_names)

# Predict using the loaded model
result = loaded_model.predict(sample_input)
print("Predicted Profit:", result[0])
```
